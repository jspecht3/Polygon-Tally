```python
import openmc
```

```python
water = openmc.Material()
water.add_element("H", 2)
water.add_element("O", 16)
water.set_density('g/cc', 1)

cell = openmc.Cell(fill = water)
```

```python
pellet_diamater = 5.5685 / 10
pin_pitch = 1.180 * pellet_diamater

pin_cell = openmc.Cell(fill = openmc.Universe(cells = [cell]))
pin_univ = openmc.Universe(cells = [pin_cell])

lattice = openmc.HexLattice()
lattice.center = (0,0)
lattice.pitch = (3**(0.5) * pin_pitch,)
```

```python
layers = []

for i in reversed(range(1,10)):
    layers.append([pin_univ] * (6*i))
layers.append([pin_univ])

lattice.universes = layers
```
