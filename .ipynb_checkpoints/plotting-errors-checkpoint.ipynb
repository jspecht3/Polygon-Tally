## Making Gifs

```python
paths = ["k_my", "k_sci"]

k_imgs_my = []
k_imgs_sci = []

imgs = [k_imgs_my, k_imgs_sci]

for i in range(len(paths)):
    # getting file paths
    for j in range(60):
        file = "output/images/{}/{}{}.png".format(paths[i], paths[i], j)
        imgs[i].append(file)

    # making gif
    save_path = "output/images/" + paths[i] + ".gif"
    print(save_path)

    frames = [Image.open(image) for image in imgs[i]]
    frame_one = frames[0]
    frame_one.save(save_path, format = "GIF",
                   append_images = frames, save_all = True,
                   duration = 200, loop = 0)
```
